```python
import pandas as pd
import numpy as np
```

```python
# read and attach the csv file, set date to be the index 
df = pd.read_csv("Desktop/Data for Seminar 1 Test.csv", header=0, index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
# daily statistics for an index

df['MERW0G1'].describe() 
```

```python
# create monthly timeframe by resampling

monthly_df = df.resample('M').mean()
print(monthly_df)
```

```python
## for each index compute the monthly return

# assign each one

index_1 = monthly_df['MERW0G1']
index_2 = monthly_df['MERG0BC']
index_3 = monthly_df['IBLUS0004']
index_4 = monthly_df['MIWO00000PUS']
index_5 = monthly_df['MIEF00000GUS']
index_6 = monthly_df['TRCCI']

# compute pct change

mr_index_1 = index_1.pct_change()
mr_index_2 = index_2.pct_change()
mr_index_3 = index_3.pct_change()
mr_index_4 = index_4.pct_change()
mr_index_5 = index_5.pct_change()
mr_index_6 = index_6.pct_change()


# print the results 

print(mr_index_1)
print(mr_index_2)
print(mr_index_3)
print(mr_index_4)
print(mr_index_5)
print(mr_index_6)
```

```python
# monthly standard dev

std_mr_index_1 = mr_index_1.std()
std_mr_index_2 = mr_index_2.std()
std_mr_index_3 = mr_index_3.std()
std_mr_index_4 = mr_index_4.std()
std_mr_index_5 = mr_index_5.std()
std_mr_index_6 = mr_index_6.std()

# print

print(std_mr_index_1)
print(std_mr_index_2)
print(std_mr_index_3)
print(std_mr_index_4)
print(std_mr_index_5)
print(std_mr_index_6)
```

```python
# create yearly timeframe by resampling

yearly_df = df.resample('Y').mean()
print(yearly_df)
```

```python
## compute annualised mean returns

index_11 = yearly_df['MERW0G1']
index_22 = yearly_df['MERG0BC']
index_33 = yearly_df['IBLUS0004']
index_44 = yearly_df['MIWO00000PUS']
index_55 = yearly_df['MIEF00000GUS']
index_66 = yearly_df['TRCCI']


# compute pct change

yr_index_11 = index_11.pct_change()
yr_index_22 = index_22.pct_change()
yr_index_33 = index_33.pct_change()
yr_index_44 = index_44.pct_change()
yr_index_55 = index_55.pct_change()
yr_index_66 = index_66.pct_change()


# print the results 

print(yr_index_11)
print(yr_index_22)
print(yr_index_33)
print(yr_index_44)
print(yr_index_55)
print(yr_index_66)
```

```python
# yearly standard dev

std_yr_index_11 = mr_index_1.std() * np.sqrt(12)
std_yr_index_22 = mr_index_2.std() * np.sqrt(12)
std_yr_index_33 = mr_index_3.std() * np.sqrt(12)
std_yr_index_44 = mr_index_4.std() * np.sqrt(12)
std_yr_index_55 = mr_index_5.std() * np.sqrt(12)
std_yr_index_66 = mr_index_6.std() * np.sqrt(12)

# print

print(std_yr_index_11)
print(std_yr_index_22)
print(std_yr_index_33)
print(std_yr_index_44)
print(std_yr_index_55)
print(std_yr_index_66)
```

```python
## equally weighted protfolio of the 8 indices

num_asset = len(df.columns)
weight = [1/num_asset for i in range(num_asset)]

num_asset
weight
```

```python
# annualised mean return

ret = df.pct_change()
ret.mul(weight, axis = 'columns').sum(axis = 1).plot()
```

```python
# historical volatility

ret.mul(weight, axis = 'columns').sum(axis = 1).std()



```
